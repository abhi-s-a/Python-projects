```python
def ddmm(ddv,mmv):
    mmj={'dd':ddv,'mm':mmv}

    if  0<mmj.get('mm')<13 and 0<mmj.get('dd')<32:
        if mmj.get('mm')<8 and mmj.get('mm')%2==0:
            if mmj.get('mm')==2:
                if mmj.get('dd')>29:
                    return ('error')
            elif mmj.get('dd')>30:
                return ('error')
            
        if mmj.get('dd')>30:
            return ('error')
        else:
            return mmj

    else:
        return ('error')
```

```python
def check_pass(pa1):
    l1=list(pa1)
    count1=0;count2=0;count3=0;count4=0
    if len(l1)>=8:

        for i in l1:
            if i.isupper():
                count1+=1
                    
            elif i.islower():
                count2+=1
                
            elif '0'<=i<='9':
                count3+=1

            else:
                count4+=1
    else:
        print(f'\nyour password has only {len(l1)} character long increase your password ')  
        return False
    D=dict(upper=count1,lower=count2,number=count3,special=count4)
    if count1==0 or count2==0 or count3==0 or count4==0:
        print(f'\nyour password has missing characters {D} \nRe-enter your password ')
        return False
    if count1>0 and count2>0 and count3>0 and count4>0:
        return pa1
```
